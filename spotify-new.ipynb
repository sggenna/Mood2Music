since the dataset had a lot of mixed genres such as (metal heavy, heavy metal) or variations of genres like punk rock and math rock i unified them into more general categories

```python
def unify_genres (genre_s):
    genre_s = genre_s.lower()
    if 'hip hop' in genre_s or 'hip-hop' in genre_s or 'rap' in genre_s:
        return 'hip_hop'
    elif 'pop' in genre_s and 'k-pop'not in genre_s:
        return 'pop'
    elif 'rock' in genre_s:
        return 'rock'
    elif 'electronic' in genre_s or 'house' in genre_s or 'electro' in genre_s or 'dance' in genre_s or 'experimental' in genre_s or 'dance' in genre_s:
        return 'electronic'
    elif 'jazz' in genre_s:
        return 'jazz'
    elif 'country' in genre_s:
        return 'country'
    elif 'folk' in genre_s:
        return 'folk'
    elif 'k-pop' in genre_s:
        return 'kpop'
    elif 'alternative' in genre_s:
        return 'alternative'
    elif 'funk' in genre_s:
        return 'funk'
    elif 'punk' in genre_s:
        return 'punk'
    elif 'soul' in genre_s:
        return 'soul'
    elif 'blues' in genre_s:
        return 'blues'
    elif 'drum and bass' in genre_s:
        return 'drum and bass'
    elif 'classical' in genre_s:
        return 'classical'
    elif 'reggae' in genre_s:
        return 'reggae'
    elif 'indie' in genre_s:
        return 'indie'
    elif 'latin' in genre_s:
        return 'latin'
    elif 'r&b' in genre_s or 'rnb' in genre_s:
        return 'r&b'
    elif 'gospel' in genre_s:
        return 'gospel'
    elif 'trap' in genre_s:
        return 'trap'
    elif 'ambient' in genre_s:
        return 'ambient'
    elif 'metal' in genre_s:
        return 'metal'
    elif 'comedy' in genre_s:
        return 'comedy'
    elif 'soundtrack' in genre_s:
        return 'soundtrack'
    elif 'hardcore' in genre_s:
        return 'hardcore'
    elif 'emo' in genre_s:
        return 'emo'
    elif 'deathcore' in genre_s:
        return 'deathcore'
    else:
        return 'other'
```

```python
df['Genre'] = df['Genre'].apply(unify_genres)
```
